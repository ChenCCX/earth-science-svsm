```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
dfes = pd.read_csv('RiverRoad_Quan_RF.csv')
dfes
```

```python
dfes_wetland = dfes[dfes['wetland']=='wetland']
dfes_nonwetland = dfes[dfes['wetland']=='nonwetland']
dfes_nonwetland
dfes_wetland
```

```python
df_corr = pd.concat([dfes[dfes.columns[1:61]], dfes['FID_RiverRd_AOI']], axis=1)
corr = df_corr.corr()
df_corr = pd.DataFrame(corr['FID_RiverRd_AOI'])
df_corr = df_corr.sort_values(by=['FID_RiverRd_AOI'])
#Remove ['FID_RiverRd_AOI'] row because corr is automatically 1
df_corr = df_corr[:-1]
df_corr
```

```python
df_corr_abs = df_corr.abs().sort_values(by=['FID_RiverRd_AOI'])
df_corr_abs
```
